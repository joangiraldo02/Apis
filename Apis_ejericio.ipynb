```python
import requests

url= "https://thesimpsonsapi.com/api/characters"

respuesta= requests.get(url)
```

```python
if respuesta.status_code == 200:
    datos= respuesta.json()
    lista_personajes=datos["results"]
    primer_personaje=lista_personajes[0]
    print("---DATOS DEL PRIMER PERSONAJE---")
    print(primer_personaje)
    if 'name' in primer_personaje:
        print(f"Nombre: {primer_personaje['name']}")
        for personaje in lista_personajes[:5]:
            print(personaje['name'])
else:
    print("Hubo un error al conectar con la API")
```

```python
import pandas as pd 

df_simpsons=pd.DataFrame(lista_personajes)

df_simpsons.head()
```
